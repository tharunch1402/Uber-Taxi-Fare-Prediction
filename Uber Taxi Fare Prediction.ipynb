```python
from scipy.stats import f_oneway

f_stat, p_value = f_oneway(
    A[A['Vehicle_Type'] == 'Mini']['Fare_Amount'],
    A[A['Vehicle_Type'] == 'Sedan']['Fare_Amount'],
    A[A['Vehicle_Type'] == 'SUV']['Fare_Amount'],
    A[A['Vehicle_Type'] == 'Auto']['Fare_Amount'],
    A[A['Vehicle_Type'] == 'Bike']['Fare_Amount']
)

print("F-statistic :", f_stat)
print("P-value :", p_value)

if p_value < 0.05:
    print("Vehicle_Type has significant relationship with Total_Fare")
else:
    print("No significant relationship")
```

```python
from scipy.stats import f_oneway

f_stat, p_value = f_oneway(
    A[A['Weather_Condition'] == 'Clear']['Fare_Amount'],
    A[A['Weather_Condition'] == 'Cloudy']['Fare_Amount'],
    A[A['Weather_Condition'] == 'Rainy']['Fare_Amount'],
    A[A['Weather_Condition'] == 'Stormy']['Fare_Amount']
)

print("F-statistic :", f_stat)
print("P-value :", p_value)

if p_value < 0.05:
    print("Vehicle_Type has significant relationship with Total_Fare")
else:
    print("No significant relationship")
```

```python
from scipy.stats import f_oneway

f_stat, p_value = f_oneway(
    A[A['City_Region'] == 'Urban']['Fare_Amount'],
    A[A['City_Region'] == 'Suburban']['Fare_Amount'],
    A[A['City_Region'] == 'Rural']['Fare_Amount']
)

print("F-statistic :", f_stat)
print("P-value :", p_value)

if p_value < 0.05:
    print("Vehicle_Type has significant relationship with Total_Fare")
else:
    print("No significant relationship")
```

## Splitting the data (Holdout Approach)

```python
X = A.drop(columns = ['Demand_In_Area','Driver_Rating','Booking_Lead_Time_min','Passenger_Count','Fare_Amount',])
y = A['Fare_Amount']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state = 23)
```

## Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder,PowerTransformer,StandardScaler
```

```python
numerical_cols = ['Distance_km', 'Hour_of_Day', 'Day_of_Week', 'Is_Peak_Hour',
       'Traffic_Density_Index', 'Estimated_Duration_min',
       'Surge_Multiplier',]
nominal_cols = [ 'Vehicle_Type','Weather_Condition',]
ordinal_cols = ['City_Region']
order = [['Urban', 'Suburban', 'Rural']]
```

```python
numerical_pipeline = Pipeline([('Imputation',SimpleImputer(strategy = 'mean')),
                               ('Scaling',StandardScaler()),
                               ('Power_transformation',PowerTransformer())])
numerical_pipeline
```

```python
ordinal_pipeline = Pipeline([('Imputation',SimpleImputer(strategy = 'most_frequent')),
                             ('Encoding',OrdinalEncoder(categories = order))])
ordinal_pipeline
```

```python
nominal_pipeline = Pipeline([('Imputation',SimpleImputer(strategy = 'most_frequent')),
                             ('Encoding',OneHotEncoder(dtype = 'int',drop = 'first',handle_unknown = 'ignore'))])
nominal_pipeline
```

## Column Transformer

```python
Data_preprocessing = ColumnTransformer([('Numerical transformation',numerical_pipeline,numerical_cols),
                                        ('Ordinal trsansformation',ordinal_pipeline,ordinal_cols),
                                        ('Nominal transformation',nominal_pipeline,nominal_cols)])
Data_preprocessing
```

```python
Data_preprocessing.fit_transform(X)
```

## Algorithms

### XGBoost

```python
xgb = Pipeline([('Data_preprocessing',Data_preprocessing),
                ('XGBoost',XGBRegressor(max_depth=6,
        learning_rate=0.05,
        n_estimators= 200,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42))])
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_test)
r2_score(y_test,y_pred)
```

### KNN

```python
lr = Pipeline([('Data Preprocessing',Data_preprocessing),
               ('Algorithm',KNeighborsRegressor())])
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
r2_score(y_test,y_pred)
```

### DecisionTree

```python
dt = Pipeline([('Data Preprocessing',Data_preprocessing),
               ('Algorithm',DecisionTreeRegressor())])
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
r2_score(y_test,y_pred)
```

## RandomForest

```python
rf = Pipeline([('Data Preprocessing',Data_preprocessing),
               ('Algorithm',RandomForestRegressor())])
rf.fit(X_train,y_train)
y_pred = rf.predict(X_test)
r2_score(y_test,y_pred)
```

# Deployement (Gradio)

```python
import pandas as pd

def prediction( Distance_km, Vehicle_Type, Hour_of_Day, Day_of_Week,
                Is_Peak_Hour, Traffic_Density_Index, Weather_Condition,
                 Estimated_Duration_min, Surge_Multiplier,
                 City_Region):

    data = pd.DataFrame([[ Distance_km, Vehicle_Type, Hour_of_Day, Day_of_Week,
                          Is_Peak_Hour, Traffic_Density_Index, Weather_Condition,
                          Estimated_Duration_min, Surge_Multiplier,
                          City_Region,]],columns=['Distance_km', 'Vehicle_Type', 'Hour_of_Day', 'Day_of_Week',
                                        'Is_Peak_Hour', 'Traffic_Density_Index', 'Weather_Condition',
                                         'Estimated_Duration_min', 'Surge_Multiplier',
                                         'City_Region'
                                        ])

    result = xgb.predict(data)
    return f"The total fare is {result} rupees."
```

```python
import gradio as gr

interface = gr.Interface(
    fn=prediction,
    inputs=[
        gr.Number(label="Distance"),
        gr.Dropdown(['Mini', 'Auto', 'Sedan', 'Bike', 'SUV'],label="Vehicle_Type"),
        gr.Number(label="Time(24H)"),
        gr.Dropdown([0,1,2,3,4,5,6],label = "Week"),
        gr.Dropdown([0,1],label = "Peak Hour"),
        gr.Number(label="Traffic Density"),
        gr.Dropdown(['Clear', 'Cloudy', 'Rainy', 'Stormy'],label = "Weather Condition"),
        gr.Number(label="Estimated_Duration_min"),
        gr.Dropdown([1. , 1.4, 1.3, 1.7],label = "Surge_Multiplier"),
        gr.Dropdown(['Urban', 'Suburban', 'Rural'],label = "City_Region")
        
        
        
    ],
    outputs="text",
    title="Taxi cab Fare Prediction"
)

interface.launch()
```
